```python
import json
import os
```

```python
source_dir = "/home/al/projects/audio-language/data/data-wiktionary/spanish/tokens/successes"
sources = os.listdir(source_dir)
sources
```

```python
unique = {}
for file in sources:
    with open(os.path.join(source_dir, file), 'r') as f:
        x = json.loads(f.read())
        for word in x:
            for part in word["parts"]:
                name = part["name"]
                if name == 'symbol':
                    print(word)
                    print("\n")
                unique[name] = True
```

```python
for thing in unique.keys():
    print('"{t}",'.format(t=thing))
```
